```python
import pandas as pd
from google.colab import files
import os
os.listdir("/content")
df=pd.read_csv("/content/Students marks (3).csv")
print("Top 10 rows of the Students marks dataset :\n")
print(df.head(10))

print("\nDataset Info :\n")
print(df.info())

print("\nMissing Values :\n")
print(df.isnull().sum())

print("\nStatistical Summary :\n")
print(df.describe())

df=df.drop_duplicates()

df['Marks2']=df['Marks2'].fillna(df['Marks2'].median())

df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])

df=df.drop(columns=['City'])
df.to_csv('Students marks_new.csv',index=True)
print("\nPrecessed data saved as Students marks_new.csv")
files.download('Students marks_new.csv')
```

```python
import seaborn as sns
import pandas as pd
import os
os.listdir('/content')
df=pd.read_csv('/content/Mental_Health_and_Social_Media_Balance_new.csv')

# Detect Outliers using IQR and store relevant statistics

# Create a copy of the dataframe to store the cleaned data
df_clean = df.copy()

# Dictionary to store limits and medians for each column
column_stats = {}

for col in['Age','Daily_Screen_Time(hrs)']:
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)
  IQR=Q3-Q1

  lower_limit=Q1-1.5*IQR
  upper_limit=Q3+1.5*IQR
  median_val = df[col].median()

  column_stats[col] = {
      "lower_limit": lower_limit,
      "upper_limit": upper_limit,
      "median": median_val
  }

  print(f"\nColumn :{col}")
  print("Lower Limit :",lower_limit)
  print("Upper Limit :",upper_limit)
  print("Outliers Count",df[(df[col]<lower_limit) | (df[col]>upper_limit)][col].count())

# Retrieve and print median values
median_Age = column_stats['Age']['median']
median_Daily_Screen_Time_hrs = column_stats['Daily_Screen_Time(hrs)']['median'] # Corrected variable name and assignment

print("Age :",median_Age)
print("Daily_Screen_Time(hrs):",median_Daily_Screen_Time_hrs)

# Replace outliers with the median for 'Age'
age_lower = column_stats['Age']['lower_limit']
age_upper = column_stats['Age']['upper_limit']
df_clean.loc[(df_clean["Age"] < age_lower) | (df_clean["Age"] > age_upper), "Age"] = median_Age

# Replace outliers with the median for 'Daily_Screen_Time(hrs)'
daily_screen_time_lower = column_stats['Daily_Screen_Time(hrs)']['lower_limit']
daily_screen_time_upper = column_stats['Daily_Screen_Time(hrs)']['upper_limit']
df_clean.loc[(df_clean["Daily_Screen_Time(hrs)"] < daily_screen_time_lower) | (df_clean["Daily_Screen_Time(hrs)"] > daily_screen_time_upper), "Daily_Screen_Time(hrs)"] = median_Daily_Screen_Time_hrs

df_clean.to_csv("Outlier_file.csv",index=False)

files.download("Outlier_file.csv")
```
