```python
import matplotlib.pyplot as plt
months=['jan','feb','mar','apr','may']
temp=[20,30,22,33,43,]
plt.plot(months,temp,marker='o',color='orange',linewidth=2)
plt.title('average monthy temp of pune (jan-may)')
plt.xlabel('months')
plt.ylabel('temp')
plt.grid('true')
plt.show()
```
